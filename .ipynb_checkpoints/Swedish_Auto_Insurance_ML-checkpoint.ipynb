# Predicting the value of Swedish auto insurance claims using machine learning techniques

Our data set contains 63 samples. 
Each sample contains two pieces of information. 
The number of claims, which will be the only feature variable in our analysis. 
The total value of the cooresponding claims for a set region in Sweden. This is measured in thousands of Swedish Kronor. It will act as our target variable. 

Desired outcome: To be able to predict, with reasonable confidence, the total value of a number of auto insurance claims for a given Swedish geographical region. I will be using various ML techniques to perform this analysis, and will be commenting on the various strengths of each technique

```python
# Python package imports.
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

First step is to read and store the data in our Machine Learning environment.
The feature, and target, data are stored in X, y respectivly. I reshaped both into 1D numpy arrays for ease of use with both Matplotlib and various sikitlearn classes. 

```python
filename = 'Swedish_Auto_Insurance.txt'
with open(filename) as f:
    data = f.readlines()
X, y = [], []
for counter in range(len(data)):
    X.append(int(data[counter].split('\t')[0]))
    y.append(float(data[counter].split('\t')[1].replace(',','.').rstrip('\n')))
X = np.array(X).reshape(-1,1)
y = np.array(y).reshape(-1,1)
```

Below is a graph showing the relationship between our feature data and the targert data.

```python
plt.rcParams["figure.figsize"] = [15, 5]
plt.plot(X, y, 'o')
plt.ylim(0, 450)
plt.xlabel('Number of Insurance Claims')
plt.ylabel('Total Value of claims in thousands SEK')
plt.grid(b=True, which='both', linestyle='-', linewidth=1)
```

This shows a strong, positive, linear relationship between the feature varible, and the target response. Our data set does contain two outliers, but they are on line with the trend of the data. The relationship is also exactly what would be expected, as the number of insurance claims increases, the total value of the insurance claims in that region also increases. 

# Model Number 1
K-nearest Neighbors Regression analysis. 


We need to split our data set into two groups, a training set and a test set. Then create a model and score the model. 
The implementation of a K-nearest Neighbors analysis is fairly simple, all that needs to be decided on is the optimal value of k. In order to figure out the optimal value of k, it is best to try many values and observe which values are yielding the best results. Below I have tested all values of k from 1 to 20. For each value I stored its training and testing results. I included a plot to show the quality of our training/testing data as a function of k. 

```python
neighbors = list(range(1,20))
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
test_scores = []
train_scores = []

for n in neighbors:
    reg = KNeighborsRegressor(n_neighbors=n)
    reg.fit(X_train, y_train)    
    test_scores.append(reg.score(X_test,y_test))
    train_scores.append(reg.score(X_train, y_train))

plt.rcParams["figure.figsize"] = [15, 5]    
plt.plot(neighbors, test_scores, '.r-')
plt.plot(neighbors, train_scores, '.b-')
plt.legend(['Test Scores', 'Training Scores'], prop={"size":15})
plt.grid(b=True, which='both', linestyle='-', linewidth=1)
plt.xlabel('Number of Neighbors')
plt.ylabel('Model Score (R^2)')
```

The above graph illistrates that we will see the best model performance when setting k=11.
It allowed us to achieve a test score of approximately .71. This is a fairly weak score, and would not be a very strong predictor. Before pursuing a different approach, to obtain a better score, observing how our model is actually behaving on new data will provide valuable insight. In order to do that, I have created an array with 1000 data points evenly spaced across the scope of our feature variable. Using that array I created predictions for every one of those data points, and then plotted those predictions alongside our existing data. 

```python
# Recreated the KNR object with 10 neighbors, and fit the model using our training data. 
reg = KNeighborsRegressor(n_neighbors=10)
reg.fit(X_train, y_train)

# Test varible, 1000 points spread across the scope of our feature variable. 
line = np.linspace(0,120,1000).reshape(-1,1)

# Plotting all of our data everything together. 
plt.plot(X_train, y_train, '^', c='b', markersize=7)
plt.plot(X_test, y_test, 'v', c='g', markersize=7)
plt.plot(line, reg.predict(line), c='r')

# Labels, Legend and other graph formating. 
plt.rcParams["figure.figsize"] = [20, 7]
plt.xlabel('Number of Insurance Claims')
plt.ylabel('Total Value of claims in thousands SEK')
plt.title('10 neighbors \n train score: {:.2f} test score: {:.2f}'.format(
    reg.score(X_train, y_train), reg.score(X_test, y_test)))
plt.legend(['Training data/target', 'Test data/target', 'Model Predictions'], loc='best', prop={"size":16})
plt.grid(b=True, which='both', linestyle='-', linewidth=1)
```

The above graph does a decent job at showing us that this model is underfitting and it also helps point to other ML options that will be able to better, Linear Regression. 

# Model Number 2
Linear Regression

I used the same train/test data split in this analysis in order to be consistent between the models. 
Because we are working with a 1 dimensional feature set, linear analysis is simply going to provide us best fit line that we can predict off of.

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train, y_train)

plt.rcParams["figure.figsize"] = [15, 5]
plt.plot(X, y, 'o', markersize=7)
plt.plot(X, lr.predict(X), c='r')

plt.xlabel('Number of Insurance Claims')
plt.ylabel('Total Value of claims in thousands SEK')
plt.grid(b=True, which='both', linestyle='-', linewidth=1)
plt.legend(['Data Set', 'Linear Model\nScore: {:.2f}'.format(lr.score(X_test, y_test))], prop={"size":18} )
```

Again we see a pretty low score, that is underfitting the data set. Because we are only working with a 1D data set, utilizing other linear techniques (Ridge, Lasso), will yield the same results that we are currently looking at. And the linear model is failing to perfom as well as the KNN (k=11) model was able to perform. This is becuase it is again underfitting the model to the complexity of the data set, and we have no way to control the model complexity. Thus we are unable to improve our score with this model.
